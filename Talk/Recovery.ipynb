```python


import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
import csv
import matplotlib.patches as mpatches
import pylab as pb
import matplotlib.ticker as mtick
from matplotlib import cm
from numpy import linspace
from csv import reader
import matplotlib.colors as colors
import seaborn as sns
import pandas as pd
import scipy.misc
from mpl_toolkits.mplot3d import Axes3D
from skimage.io import imread
from matplotlib import cm as cm

```

```python
gate = []
pre = []
post = []
evo = []


with open('stability.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        gate.append(float(row[0]))
        pre.append(float(row[1]))
        post.append(float(row[2]))
        evo.append(float(row[3]))
```

```python


fig = plt.figure(figsize=(4.5,4.5))
fig.canvas.draw()

sns.set_style('ticks')

ax1 = fig.add_subplot(111, axisbg="white")
ax1.plot(gate,pre , '-', ms=10, color = 'black',  lw=3, alpha=1, mfc='#036564', dash_capstyle = 'butt', mew = '1', 
              label = 'Pre-treatment' ) 
ax1.plot(gate,post , '-', ms=10, color = 'r',  lw=3, alpha=1, mfc='red', dash_capstyle = 'butt', mew = '1', 
              label = '6 sec (fresh)' ) 
ax1.plot(gate,evo , '-', ms=10, color = 'blue',  lw=3, alpha=1, mfc='red', dash_capstyle = 'butt', mew = '1', 
              label = '6 sec (after night in UHV)' ) 


ax1.grid('off')
ax1.set_title('(d)')



ax1.set_ylabel('$I_{DS}$ (A)', fontsize = 12)
ax1.set_xlabel("V$_{GS}$ (V)", fontsize = 12)



plt.ticklabel_format(style='sci', axis='y')
#plt.ticklabel_format(style='sci', axis='x')
ax1.set_yscale('log')
#ax2.set_yscale('log')
#ax3.set_yscale('log')



#ax1.set_ylim([0.01,100])
#ax2.set_ylim([0.00000001,0.00001])

ax1.legend(loc = 'lower right', ncol = 1, fontsize=10)

plt.savefig("stability_jupster",bbox_inches='tight', dpi = 300)


plt.show()

```
